## Loser Detection

We have to make sure there are no losers at our wedding. (More accurately: We need to understand who we need more information about to make this hare-brained scheme work.)

We remove people who are unlikely to have many - if any - connections for two main reasons.
* They act as outliers on the plot and squash the interesting bit of data in the middle
* They will never be fully matched and therefore we don't want to algorithms developed to spend a lot of time pandering to them.

This is outputted and we do need to act on this information - but the solution to this problem involves talking to people and Python does not have a library for that --- *yet*.

For our purposes, a party is a loser if we only understand one interesting thing about them - even if it would connect them to multiple people.

```python
def matches_containing_party(G, party):
    return [match for match in G.edges(data="why") if party in match]

def reasons_for_matches(G, party):
    return set([match[2] for match in matches_containing_party(G, party)])

#A loser is anyone who has only one thing interesting about them
def losers(G):
    return [party for party in G.nodes if len(reasons_for_matches(G, party)) < 2]
    
def remove_losers(G):
    print("Removing %s" % losers(G))
    G.remove_nodes_from(losers(G))

remove_losers(graph)

plot_graph(graph)
plt.show()
```

Support functions. These are largely uninteresting.

```python
#It cannot be added if the reason overlaps with any existing reason two parties are connected
def match_can_be_added(matching_graph, match):
    reasons_for_start = reasons_for_matches(matching_graph, match[0])
    reasons_for_end = reasons_for_matches(matching_graph, match[1])
    
    return (match[2] not in reasons_for_start) & (match[2] not in reasons_for_end)

def potential_matches_for_party(matching_graph, connection_graph, party):
    return [match for match in matches_containing_party(connection_graph, party) if match_can_be_added(matching_graph, match)]

def least_matched_parties_with_outstanding_matches(matching_graph, connection_graph):
    parties_with_outstanding_matches = [party for party in connection_graph.nodes if len(potential_matches_for_party(matching_graph, connection_graph, party)) > 0]
    lowest_degree = min([degree for (node, degree) in matching_graph.degree if node in parties_with_outstanding_matches])
    return [party for (party, degree) in matching_graph.degree if (party in parties_with_outstanding_matches) & (degree == lowest_degree)]

def other_end_of_match(match, party):
    if(match[0] == party):
        return match[1]
    else:
        return match[0]
```

## The Algorithm

The algorithm will iteratively move matches from the raw "Connections Graph" where we store the original connection to the "Matching Graph". The ulimtate aim is to have every node in the matching graph be of degree 2 with every connection/match/edge being for a different facet of a person. That's a lot of work so getting close is good enough.

The approach:
* Given our Connections Graph G = (V, E).
* __Step 1__: Construct a blank matching graph S = (V, {})
* __Step 2__: For each 'forced' match, remove that match from E and add it to the edge set of S.
* __Step 3__: Determine whether we are ready to terminate - the condition for this is that every party with less than two matches (degree < 2 on the matching graph S) cannot be matched further (for party P, all edges (P,Q) on the connections graph G would lead to a match for a duplicate reason for either party P or Q).
* __Step 4__: If we did not terminate, we migrate an edge from the connections graph to the matching graph.
    * __Step 4a__: Select a party to match. From among the nodes on the matching graph with the lowest degree; select the one that also has the lowest degree on the connections graph. If there is a tie, choose arbitrarily.
    * __Step 4b__: Select a match adjacent to that party. From among the matches (P, Q) choose the one whose destination Q has the lowest degree on the matching graph. If there is a tie, choose arbitrarily
    * __Step 4c__: Remove that edge from the edge set of G. Add it to the edgeset of S.
* __Step 5__: Return to __Step 3__ to determine whether we terminate again.

```python
# Step 2 of Algorithm
def add_forced_matches(matching_graph, connection_graph):
    edges = [(edge[0], edge[1], {"why": edge[2]["why"]}) for edge in connection_graph.edges(data=True) if edge[2]["forced"]]
    matching_graph.add_edges_from(edges)
    fresh_graph.remove_edges_from(edges)

#Step 3 of Algorithm
def unmatched_people_have_further_mappings(matching_graph, connection_graph):
    unmatched_parties = [party for (party, degree) in matching_graph.degree if degree < 2]
    return any([len(potential_matches_for_party(matching_graph, connection_graph, party)) > 0 for party in unmatched_parties])

# Step 4a of Algorithm
#Of all parties with the least existing matches, select the most isolated one
def next_party_to_match(matching_graph, connection_graph):
    most_available_parties = least_matched_parties_with_outstanding_matches(matching_graph, connection_graph)
    outstanding_match_degrees = connection_graph.degree
    
    parties_with_degrees = [(party, outstanding_match_degrees[party]) for party in most_available_parties]
    
    lowest_degree = min([d[1] for d in parties_with_degrees])
    candidates = [party for party in parties_with_degrees if party[1] == lowest_degree]
    return candidates[0][0] #select an arbitrary party

#Step 4b of Algorithm
#Select the match that is with the most isolated party
def next_match_for_party(matching_graph, connection_graph, party):
    all_matches_for_party = potential_matches_for_party(matching_graph, connection_graph, party)
    
    existing_matchings = matching_graph.degree
    matches_with_destination_degrees = [(edge, existing_matchings[other_end_of_match(edge, party)]) for edge in all_matches_for_party]
    least_degree = min([d[1] for d in matches_with_destination_degrees])
    
    candidates = [match for match in matches_with_destination_degrees if match[1] == least_degree]
    selected_match = candidates[0][0] #select an arbitrary match
    return (selected_match[0], selected_match[1], {"why": selected_match[2]})

#Step 4c of Algorithm
def perform_match(matching_graph, connection_graph, match):
    matching_graph.add_edges_from([match])
    fresh_graph.remove_edges_from([match])
```

```python
#Execute the algorithm and plot the accompanying matching graph
fresh_graph = generate_full_graph()
remove_losers(fresh_graph)

matching_graph = nx.Graph()
matching_graph.add_nodes_from(fresh_graph.nodes)
add_forced_matches(matching_graph, fresh_graph)

while unmatched_people_have_further_mappings(matching_graph, fresh_graph):
    party = next_party_to_match(matching_graph, fresh_graph)
    match = next_match_for_party(matching_graph, fresh_graph, party)
    perform_match(matching_graph, fresh_graph, match)
        
plot_graph(matching_graph)
plt.show()
```
